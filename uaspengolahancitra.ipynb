```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar
image_path = 'image1.jpg'  # Gantilah dengan path gambar Anda
image = cv2.imread(image_path)

# Mengubah gambar ke grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 1. Histogram Equalization (Peningkatan Kontras)
equalized_image = cv2.equalizeHist(gray_image)

# 2. Sharpening (Penajaman Gambar)
kernel_sharpen = np.array([[-1, -1, -1],
                           [-1,  9, -1],
                           [-1, -1, -1]])
sharpened_image = cv2.filter2D(image, -1, kernel_sharpen)

# 3. Noise Reduction (Pengurangan Noise)
denoised_image = cv2.medianBlur(image, 5)

# Menampilkan hasil
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title("Equalized Image")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title("Sharpened Image")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(denoised_image, cv2.COLOR_BGR2RGB))
plt.title("Denoised Image")
plt.axis('off')

plt.show()
```
